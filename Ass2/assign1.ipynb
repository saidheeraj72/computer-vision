```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the captured image
gray_image = cv2.imread('img.jpeg',cv2.IMREAD_GRAYSCALE)
plt.imshow(gray_image)
```

```python
#using canny
edges_canny = cv2.Canny(gray_image, 50, 150)
```

```python
#using marr hildeberth
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
edges_marr_hildreth = cv2.Laplacian(blurred_image, cv2.CV_64F)
```

```python
c1=float(input("enter the c value for marr hildreth sharpening\n"))
sharpened_marr_hildreth = gray_image.astype(float) + c1 * edges_marr_hildreth
sharpened_marr_hildreth = np.clip(sharpened_marr_hildreth, 0, 255).astype(np.uint8)
cv2.imshow('Original', gray_image)
cv2.imshow('Sharpened with Marr Hildreth', sharpened_marr_hildreth)
cv2.waitKey(0)
cv2.destroyAllWindows()  
```
